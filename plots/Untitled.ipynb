```python
import torchaudio
import pathlib
import librosa
import librosa.display
from matplotlib import pyplot as plt
```

```python
path_data = pathlib.Path("audiofiles")

audio_samples = {_.name.split(".")[0]: torchaudio.load(_) for _ in path_data.iterdir()}
```

```python
melspec = torchaudio.transforms.MelSpectrogram(
                    sample_rate = 48000,
                    n_fft = 1024,
                    win_length = 512,
                    hop_length = 512,
                    n_mels = 64,
                    normalized = True
)

amp_to_db = torchaudio.transforms.AmplitudeToDB(stype = "power", top_db = 80)
```

```python
for name in list(audio_samples.keys()):
    sr = audio_samples[name][1]
    y = audio_samples[name][0][0]
    plt.figure(0)
    plt.title(f"Signal Wave: {name}")
    plt.plot(y)
    plt.savefig(f"{name}_signal_wave.png")
    y_melspec = melspec(y)
    y_db_spec =amp_to_db(y_melspec) 
    plt.figure(0)
    plt.title(f"Melspec: {name}")
    plt.imshow(y_melspec[:, :200])
    plt.savefig(f"{name}_melspec.png")
    plt.figure(0)
    plt.title(f"DB Melspec: {name}")
    plt.imshow(y_db_spec[:, :200])
    plt.savefig(f"{name}_db_melspec.png")
```
